```python
class Neuron:

    def __init__(self, w, f = lambda x: x):
        #YOUR CODE HERE
        self.w = w
        self.f = f

    def forward(self, x):
        #YOUR CODE HERE
        self.x = x
        return self.f(sum([w*x for w,x in zip(self.w, self.x)]))

    def backlog(self):
        #YOUR CODE HERE
        return self.x
```
